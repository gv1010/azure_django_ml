```python
import pandas as pd
df = pd.read_csv("winequality-white.csv",sep=';')
df
```

```python
df.isna().sum()
```

```python
import pandas as pd
import numpy as np

rs = np.random.RandomState(0)
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
from sklearn.preprocessing import StandardScaler
data = [[0, 0], [0, 0], [1, 1], [1, 1]]
scaler = StandardScaler()
print(scaler.fit(data))
```

```python
X, y = df.iloc[:,:-1], df.iloc[:,-1]
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
scaler = StandardScaler()
x = scaler.fit_transform(X_train)
```

```python
import joblib
joblib.dump(scaler, 'scaler.joblib')  
```

```python
model = joblib.load('scaler.joblib')
x = model.fit_transform(X_train)
```

```python
from sklearn.svm import SVR
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np
regr = SVR(C=1.0, epsilon=0.2)
regr.fit(x, y_train)
```

```python
x_t = scaler.transform(X_test)
x_t
```

```python
x_t = scaler.transform(X_test)
regr.predict(x_t)
```

```python
import joblib
joblib.dump(regr, 'model.joblib')  
```

```python
model = joblib.load('model.joblib')
model.predict(x_t)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_squared_error(regr.predict(X_test), y_test)
```

```python
mean_absolute_error(regr.predict(X_test), y_test)
```

```python
import pickle 
  
# Save the trained model as a pickle string. 
saved_model = pickle.dumps(regr) 
```

```python
import joblib 
  
# Save the model as a pickle in a file 
joblib.dump(saved_model, 'svmdj.pkl') 
  
# Load the model from the file 
load_model = joblib.load('svmdj.pkl')
```

```python
load_model.predict(X_test)
```

```python
filename = 'finalized_model.sav'
pickle.dump(regr, open(filename, 'wb'))
loaded_model = pickle.load(open(filename, 'rb'))
loaded_model.predict(X_test)
```
